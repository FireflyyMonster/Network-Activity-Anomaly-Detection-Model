#### We have 3 categorical datatype columns, Hence dealing with categorical data as follows:
The Protocol Type column has just 3 categories hence, using One Hot Encoding 

```python
df.protocoltype
```

```python
df.protocoltype.value_counts()
```

```python
df=pd.get_dummies(df,columns=['protocoltype'], prefix=['protocol'])  # Since their are only 3 unique values in this column, One hot encoding is used
```

```python
df["protocol_icmp"]=df["protocol_icmp"].map({True:1,False:0}) # Mapping the boolean values to 1 and 0 respectively
df["protocol_tcp"]=df["protocol_tcp"].map({True:1,False:0})
df["protocol_udp"]=df["protocol_udp"].map({True:1,False:0})
```

```python
df
```

```python
df.shape # no of columns has increased
```

```python
df.service.value_counts()
```

```python
df.service.nunique()
```

 Here there are 65 unique values hence using Mean Ordinal Encoding

```python
Mean_of_services= df.groupby(['service'])['attack'].mean().to_dict()
# Calculating the mean of the Target variable according to the Service and converting it to a Dictionary to fit into the map function
```

```python
Mean_of_services
```

```python
df['service'] =df['service'].map(Mean_of_services)
```

```python
df['service']
```

```python
df.flag.value_counts()
```

Using Label Encoding since their are just 10 unique values

```python
flag_convert={"SF":1,"S0":2,"REJ":3,"RSTO":4,'S1':5, "RSTR":6,'S2':7,'S3':8,"OTH":9,'SH':10}
```

```python
df['flag']=df.flag.map(flag_convert)
```

```python
import sklearn 
import numpy as np
```

```python
target = df.loc[:,['attack']]
```

```python
training_columns=df.loc[:,['duration', 'service', 'flag', 'srcbytes', 'dstbytes', 'land',
       'wrongfragment', 'urgent', 'hot', 'numfailedlogins', 'loggedin',
       'numcompromised', 'rootshell', 'suattempted', 'numroot',
       'numfilecreations', 'numshells', 'numaccessfiles', 'numoutboundcmds',
       'ishostlogin', 'isguestlogin', 'count', 'srvcount', 'serrorrate',
       'srvserrorrate', 'rerrorrate', 'srvrerrorrate', 'samesrvrate',
       'diffsrvrate', 'srvdiffhostrate', 'dsthostcount', 'dsthostsrvcount',
       'dsthostsamesrvrate', 'dsthostdiffsrvrate', 'dsthostsamesrcportrate',
       'dsthostsrvdiffhostrate', 'dsthostserrorrate', 'dsthostsrvserrorrate',
       'dsthostrerrorrate', 'dsthostsrvrerrorrate', 'lastflag',
       'protocol_icmp', 'protocol_tcp', 'protocol_udp']]
```

```python
training_columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
training_column_train, val_training_column, target_train,val_target=train_test_split(training_columns,target,test_size = 0.3,random_state = 2)
target_train = np.ravel(target_train)
val_target= np.ravel(val_target)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler(feature_range = (0,1))
```

```python
scaled_train_columns = scaler.fit_transform(training_column_train)
```

```python
scaled_train_columns
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel= LogisticRegression()
```

```python
logmodel.fit(scaled_train_columns,target_train)
```

```python
val_training_scaled = scaler.transform(val_training_column)
```

```python
logmodel.predict(val_training_scaled)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(logmodel.predict(val_training_scaled), val_target)
```

```python
logmodel.score(scaled_train_columns,target_train)
```

```python
logmodel.score(val_training_scaled,val_target)
```
